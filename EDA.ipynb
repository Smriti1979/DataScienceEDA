 Calculating the average happiness score for each year

```python
df17['Year'] = 2017
df18['Year'] = 2018
df19['Year'] = 2019

combined_df = pd.concat([df17, df18, df19])
average_happiness_by_year = combined_df.groupby('Year')['Happiness.Score'].mean()
print(average_happiness_by_year)
```

```python
df15.rename(columns={'Economy (GDP per Capita)':'GDP per capita'},inplace=True)
data2=df15.filter(['Country','GDP per capita',"Year"],axis=1)

df16.rename(columns={'Economy (GDP per Capita)':'GDP per capita'},inplace=True)
data3=df16.filter(['Country','GDP per capita',"Year"],axis=1)

df17.rename(columns={'Economy..GDP.per.Capita.':'GDP per capita'},inplace=True)
data4=df17.filter(['Country','GDP per capita','Year'],axis=1)

df18.rename(columns={'Country or region':'Country'},inplace=True)
data5=df18.filter(['Country','GDP per capita',"Year"],axis=1)

data2=pd.concat([data3,data4,data5,data2])
```

```python
plt.figure(figsize=(10, 10))
h=pd.pivot_table(data2,values='GDP per capita',
index=['Country'],
columns='Year')
h=h.head(10)
sns.heatmap(h,cmap="RdBu")
```

```python
plt.figure(figsize=(15, 5))
sns.barplot(x=df17["Country"].head(100), y=df17["Happiness.Score"].head(100))
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('Happiness Rate')
plt.title('Happiness rate of top 100 countries')
plt.show()
```

 Visualizing the happiness rate of only the top 10 countries

```python
happiness_by_region = df17.groupby('Country')['Happiness.Score'].mean()


top_10_regions = happiness_by_region.nlargest(10).index
filtered_data = df17[df17['Country'].isin(top_10_regions)]
happiness_by_region_top10 = filtered_data.groupby('Country')['Happiness.Score'].mean()

labels = happiness_by_region_top10.index
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'orange', 'green', 'blue', 'red', 'purple', 'pink']


plt.figure(figsize=(7, 7))
plt.pie(happiness_by_region_top10, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Happiness Rate of Top 10 Regions")
plt.show()
```
